```python
# Homework 1
# Write a program to plot the funciton from -pi to pi.
# Choose n=11, 21 and 31 and plot all these three curves in one figure. Please use numpy library and matplotlib.

# These commands import all necessary libraries.
from numpy import *
import math
import matplotlib.pyplot as plt

# sets up x as the horizontal axis
x = linspace(-math.pi, math.pi)

# Introduces all functions: n=11, n=21, and n=31.
# n=11
y1 = (math.pi / 2) + 2*(sin(x) + (sin(3*x)/3) + (sin(5*x)/5) + (sin(7*x)/7) + (sin(9*x)/9) + (sin(11*x)/11))
# n=21
y2 = (math.pi / 2) + 2*(sin(x) + (sin(3*x)/3) + (sin(5*x)/5) + (sin(7*x)/7) + (sin(9*x)/9) + (sin(11*x)/11) + (sin(13*x)/13) + (sin(15*x)/15) + (sin(17*x)/17) + (sin(19*x)/19) + (sin(21*x)/21))
# n=31
y3 = (math.pi / 2) + 2*(sin(x) + (sin(3*x)/3) + (sin(5*x)/5) + (sin(7*x)/7) + (sin(9*x)/9) + (sin(11*x)/11) + (sin(13*x)/13) + (sin(15*x)/15) + (sin(17*x)/17) + (sin(19*x)/19) + (sin(21*x)/21) + (sin(23*x)/23) + (sin(25*x)/25) + (sin(27*x)/27) + (sin(29*x)/29) + (sin(31*x)/31))

# Plots all functions on the same graph
plt.plot(x, y1, 'r') # plotting y1 as red 
plt.plot(x, y2, 'b') # plotting y2 as blue 
plt.plot(x, y3, 'g') # plotting y3 as green
plt.show()
```
